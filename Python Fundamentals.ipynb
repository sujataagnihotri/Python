# Variables

**NOTES:**
1. Variables are the containers for the future information.
2. **Dynamic Typing :** Where you do not need to declare a variable and mention the type of variable. Python interpreter will identify it smartly. (E.g. x=5)
3. **Static Typing :** Where we need to declare a variable and mention the type of variable (E.g. int x=5)
4. **Dynamic Binding :** In python, when we create a variable, the data type of that variable is not fixed; we can change the same within the same program.(a=5 and lateron we can also mention a='Sujata')
5. **Static Binding :** When we create a variable, the data type of that variable is fixed; we cannot change the same within the same program.

# Operators

**NOTE: Types of Operators**
1. Arithmetric Operators
2. Relational Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Membership Operators

```python
# Arithmetric Operators
print(10+4) # Add
print(10-4) # Substract
print(10*4) # Multiply
print(10/4) # Divide
print(10//4) # Integer Division - It will return only integer from the result if the result is float. E.g if Result will be 3.5 then final result will be 3
print(10%4) # Modulus Operator - It will return the remainder from the result
print(10**4) # Power of E.g. 10^4 = 10*10*10*10
```

```python
# Relational Operators
print(5>4)
print(5<4)
print(5>=4)
print(5<=4)
print(5==4) # Double Equal to operator to check whether both numbers are equal or not
print(5!=4) # Not Equal to
```

```python
# Logical Operators
print (1 and 0)
print (1 or 0)
print (not 1)
print (not 0)
```

```python
# Bitwise Operators
print (10 & 3) # Bitwise AND >> 1010 & 0011 = 0010
print (10 | 3) # Bitwise OR >> 1010 | 0011 = 1011
print (10 ^ 3) # Bitwise XOR >> 1010 ^ 0011 = 1001  (If same then "0" & for Different then "1")
print (~10) # Bitwise Not - Reverse 1 to 0 and vise versa to find the result
print (5 >> 2) # Bitwise Left - Move bits by mentioned numbers
print (5 << 2) # Bitwise Right - Move bits by mentioned numbers
```

```python
# Assignment Operators
# E.g. a = 2; here "=" is used as Assignment operator
a=2
a+=3 # a=a+3
print(a)
```

```python
# Membership Operators
# 1. In operator
print ('s' in 'Sujata')
print ('S' in 'Sujata')
print (1 in [1,2,3,4,5])
print (1 in (1,2,3,4,5))
print (1 in {1,2,3,4,5})


# 2. Not In Operator
print ('R' not in 'Sujata')
```
